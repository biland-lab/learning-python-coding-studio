```python

"""
Algoritma pseudocode
(kode yang seperti bahasa manusia)
bubble sort:
listawal : [5, 3, 8, 4, 2] 
mulai
pjglist = len(list awal)
untuk index dari 0 ke pjglist-1
    untuk jindex dari 0 ke pjglist-index-1:
           if arr[jindex] > arr[jindex+1]:
                tukar arr[jindex] dan arr[jindex + 1]
selesai"""


def bubble_sort(data_list):
     #mulai
    pjglist = len(data_list)  #ambil panjang list
    #lakukan looping sepanjang list
    for index in range(0, pjglist - 1):
        #lakukan looping di dalam list untuk mengecek besar kecil data
        for jindex in range(0, pjglist - index - 1):
              #bandingan data yang bersebelahan
              if data_list[jindex] > data_list[jindex + 1]:
                #swap/ tuker posisinya
                data_list [jindex], data_list[jindex + 1] = data_list[jindex +1], data_list[jindex]
                """
                data_list[jindex] - data_list[jindex + 1]
                data_list[jindex + 1] - data_list[jindex + 1]
                """
    return data_list

#cara menggunakan function
list_nilai  = [90, 60, 70, 56, 89]
print("List Nilai Awal: ", list_nilai)
print("List Nilai Terurut: ", bubble_sort(list_nilai))
```
