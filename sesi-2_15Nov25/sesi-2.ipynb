```python
#control flow

# Operator Aritmetika
nilai_a = 90
nilai_b = 100

print(nilai_a + nilai_b) #operator +
print(nilai_a - nilai_b) #operator -
print(nilai_a / nilai_b) #operator / (bagi)
print(nilai_a * nilai_b) #operator x (kali)
print(nilai_a ** 2) #operator kuadrat
print(nilai_a % nilai_b) #operator modulus / sisa hasil bagi
```

```python
#operator comparison (hasil true/ false)
# <, >, <=, >=, ==, != (tidak sama dengan)
a = 10
b = 5

print (a>b) #true
print (a >= b) #True
print (a <= b) #False
print (a == b) #false
print (a != b) #True
```
