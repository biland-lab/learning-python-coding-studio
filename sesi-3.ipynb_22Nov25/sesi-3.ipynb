```python
# Set -> himpunan -> {}
# tidak bisa menerima duplicate elemen
# tidak ada urutan
# 
himpunan_nama = {"gio", "fadlan", "clairine"}
print ("set awal: ", himpunan_nama)
# menambah data ke dalam set -> add()
# himpunan_nama.add("anton")
# print ("set update: ", himpunan_nama)

# #menghapus data dalam set
# #method remove()
himpunan_nama.remove("fadlan") #error baris kode di bawah tidak akan dijalankan
print ("set update setelah diremove: ", himpunan_nama)

# method discard()
himpunan_nama.discard("ANTON")
print("set update setelah di remove: ", himpunan_nama)


# Gabungan/ join, selisih, irisan/ intersection
himp_siswa_a = {"clairine", "fadlan", "gio"}
himp_siswa_b = {"fadlan", "hafiz", "anton"}
                
# gabungan/ join -> |
print ("gabungan kedua himp: ", himp_siswa_a | himp_siswa_b)

# selisih -> -
print("selisih himp a dan himp b: ", himp_siswa_a - himp_siswa_b)
print("selisih himp b dan himp a: ", himp_siswa_b - himp_siswa_a)

# irisan -> &
print ("irisan himpunan: ", himp_siswa_a & himp_siswa_b)
```

```python
#Dictionary -> kamus {key:value}

kontak_telepon = {
    "clairine": 998,
    "fadlan": 997,
    "gio": 996,
}
print ("dictionary awal: ", kontak_telepon)
# menambahkan data ke dalam dictionary -> indexing
kontak_telepon["hafiz"] = 888
print ("dictionary update: ", kontak_telepon)

# merubah data dengan indexing
kontak_telepon ["gio"] = 889
print ("dictionary update: ", kontak_telepon)

# menghapus data dalam dict
# pop()
kontak_telepon.pop("gio")
print("dictionary update: ", kontak_telepon)

# del
del kontak_telepon["fadlan"]
print("dictionary update: ", kontak_telepon)

# keys(), values(), items()
#keys() -> mengambil semua key()
print(kontak_telepon.keys())

# values() -> mengambil seua value
print(kontak_telepon.values())

# items() -> mengambil semua key dan value
print(kontak_telepon.items())
```

```python
# Chatbot Sederhana
import random

#library random
# Respon pertanyaan: Suatu program yang bisa merespon pertanyaan, disimpan dalam key (Q) dan value (jwbn)
# key: value (pertanyaan:jawaban)


respons = {
        "hai": ["halo apa kabar?", "Halo", "Selamat Datang"],
        "nama kamu siapa": ["Saya Chatbot Gacor", "Saya Pembantu Anda"],
        "bye": ["Selamat Tinggal", "Sampai Jumpa", "Bye"]
}


def chatbot():
    #minta input (user input)
    #cek kondisi (inputan = data di respons)

    #perulangan untuk mengecek kondisi
    while True: #ini kondisi benar (akan eksekusi)
        user_input =  input ("üó£Ô∏èUser: ").lower()

        # cek user input apakah ada di respons
        if user_input in respons:
            print("ü§ñ Bot: ", random.choice(respons[user_input]))
        elif user_input == "keluar":
            print("ü§ñ Bot: Sampai Jumpa")
            break
        else:
            print(" Bot: Maaf Saya Tidak Mengerti")
#panggil function
chatbot()
```
